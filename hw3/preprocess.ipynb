```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
import glob
import os
import numpy as np
from PIL import Image
import torchvision.models as models
import copy
from torchvision.utils import save_image
import PIL
import skimage.io
import copy
from scipy import stats
import multiprocessing as mp
```

```python
def getImg(fn):
    if fnToIndex(fn) % 1000 == 2:
        print(fn)
    trans = transforms.ToTensor()
    return trans(Image.open(fn)).numpy()

def fnToIndex(fn):
    return int(fn.split('/')[-1].split('.')[0])
```

```python
def loadAndSaveImage(loadRoot, saveRoot):
    fnList = sorted(glob.glob(os.path.join(loadRoot,'*.png')))
    pool = mp.Pool(mp.cpu_count())
    imgList = pool.map(getImg, fnList)
    for i, img in enumerate(imgList):
        np.save(saveRoot+'/'+str(i)+".npy", img)
    print('data saving completed at',saveRoot)
```

```python
loadAndSaveImage('hw3_data/digits/mnistm/train', 'hw3_data/p3_npy/mnistm/train')
loadAndSaveImage('hw3_data/digits/mnistm/test', 'hw3_data/p3_npy/mnistm/test')
loadAndSaveImage('hw3_data/digits/svhn/train', 'hw3_data/p3_npy/svhn/train')
loadAndSaveImage('hw3_data/digits/svhn/test', 'hw3_data/p3_npy/svhn/test')
loadAndSaveImage('hw3_data/digits/usps/train', 'hw3_data/p3_npy/usps/train')
loadAndSaveImage('hw3_data/digits/usps/test', 'hw3_data/p3_npy/usps/test')
```
